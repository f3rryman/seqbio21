```python
import matplotlib.pyplot as plt

names = ["short", "medium", "long"]
modes = ["Needleman-Wunsch", "Hirschberger", "NW without table"]

timeNW = [0.027566194534301758, 0.20958232879638672, 8.23521113395691]
bytesNW = [36669, 195637, 6409790]

timeHirsch = [0.0239102840423584, 0.12775802612304688, 5.154118776321411]
bytesHirsch = [26400, 113284, 3307522]

timeNWwot = [0, 0, 0]
bytesNWwot = [0, 0, 0]

smallsize = 16
bigsize = 20

ytmax = 9
ybmax = 7000000

plt.figure(figsize=(16, 6))
plt.suptitle("Time consumption for alignment calculation", size=bigsize)
plt.rc("font", size=smallsize)

plt.subplot(131)
plt.bar(names, timeNW)
plt.axis([None, None, None, ytmax])
plt.title(modes[0])
plt.ylabel("time [1/s]")

plt.subplot(132)
plt.bar(names, timeHirsch)
plt.axis([None, None, None, ytmax])
plt.title(modes[1])

plt.subplot(133)
plt.bar(names, timeNWwot)
plt.axis([None, None, None, ytmax])
plt.title(modes[2])



plt.figure(figsize=(16, 6))
plt.suptitle("Peak memory usage for alignment calculation", size=bigsize)

plt.subplot(131)
plt.bar(names, bytesNW)
plt.axis([None, None, None, ybmax])
plt.title(modes[0], size="16")
plt.ylabel("peak memory [1/bytes]", size="16")

plt.subplot(132)
plt.bar(names, bytesHirsch)
plt.axis([None, None, None, ybmax])
plt.title(modes[1], size="16")

plt.subplot(133)
plt.bar(names, bytesNWwot)
plt.axis([None, None, None, ybmax])
plt.title(modes[2], size="16")

```
